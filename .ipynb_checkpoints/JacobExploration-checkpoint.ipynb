```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.pyplot import figure
import seaborn as sns # for visualiation
```

```python

listings = pd.read_csv("data/listings.csv", low_memory = False)
amenities = pd.read_csv("data/with_amenities.csv", low_memory = False)
```

```python
listings['weekly_price'] = listings['weekly_price'].replace({'\$': '', ',': ''}, regex=True)
listings['weekly_price'] = listings['weekly_price'].dropna(axis=0, how="any")
listings['weekly_price'] = listings['weekly_price'].astype(float)
```

```python
listings['review_scores_value'] = listings['review_scores_communication'].dropna(axis=0, how="any")
```

```python

corr = listings.corr().abs()
s = corr.unstack()
# sort from greatest to least correlation coeffcient 
sof = s.sort_values(kind="quicksort", ascending = False)
# top correlations related to weekly price
sof['weekly_price']
```

```python
plt.subplots(figsize=(12,9))

sns.violinplot(x="bedrooms", y="weekly_price", data=listings)
```

```python
plt.subplots(figsize=(12,9))

sns.violinplot(x="beds", y="weekly_price", data=listings)
```

```python
plt.subplots(figsize=(12,9))

sns.violinplot(x="accommodates", y="weekly_price", data=listings)
```

```python


fig,ax = plt.subplots(2,2,figsize=(15,15))

ax[0,0].scatter(listings.loc[:,'square_feet'], listings.loc[:,'weekly_price'], alpha = 0.6)
ax[0,0].set_xlabel('square_feet')
ax[0,0].set_ylabel('weekly_price')
ax[0,0].set_title('Number of Bedrooms and Weekly Price')

ax[0,1].scatter(listings.loc[:,'guests_included'], listings.loc[:,'weekly_price'], alpha = 0.3, color = 'orange')
ax[0,1].set_xlabel('guests_included')
ax[0,1].set_ylabel('weekly_price')
ax[0,1].set_title('Number of Bedrooms and Weekly Price')
```
